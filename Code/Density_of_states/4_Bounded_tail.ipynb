## Generalized

```python
from openpyxl import load_workbook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style # import style module
from scipy import stats
import math
from scipy.integrate import quad


a = 10
B = 3/2

N = 10
an = a 
bn = a*N**(-1/B)



x = np.linspace(0,10,100)

#1) Define parent function
def p(x):
    if x < a and x > 0:
        p = (B*a**(-B))*((a-x)**(B-1)) 
    else:
        p = 0
    return p


#p_a = [p(i) for i in x]
#plt.plot(x,p_a)

#2) Define pmax
def P(x):
    return quad(p,0,x)[0]

#P_a = [P(i) for i in x]
#plt.plot(x,P_a)

def pmax(x,N):
    pmax = N*p(x)*((P(x))**(N-1))
    return pmax

#print(pmax(7))    
#pmaxa = [bn*pmax(i) for i in x]    
#plt.plot(x,pmaxa)


r = np.linspace(0,10,100)


    
def Integrand(x,r,N):
    return p(x-r)*pmax(x,N-1)


#Int_a = [Integrand(i,9) for i in x]
#plt.plot(x,Int_a)


def pf(r):
    if r < a and r > 0:
        p = (B*a**(-B))*((r)**(B-1)) 
    else:
        p = 0
    return p


pf_a = [pf(i) for i in r]


##4) Plotting everything
plt.figure(figsize=(12, 8))
plt.grid()

rho_a_1  = [quad(Integrand, i , np.inf, args=(i,10))[0] for i in r]
rho_a_2  = [quad(Integrand, i , np.inf, args=(i,100))[0] for i in r]


plt.plot(r,rho_a_1, label = "10")
plt.plot(r,rho_a_2, label = "100")
plt.plot(r, pf_a, label = "limit")

plt.ylabel(r'  $ \ \  \overline{\rho (r, N)}  \ \  $ ',fontsize=16)
plt.xlabel(r'  $\ \ r  \ \  $ ', fontsize=16)
plt.legend(fontsize=10)


plt.savefig("Part3.png")



    


```
