```python
from openpyxl import load_workbook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style # import style module
from scipy import stats
import math
from scipy.integrate import quad

a = 5
N = 100000
an = 0
bn = N**(1/a)

x = np.linspace(0.000000000001,5*bn , 100 )

#1) Define parent function
def p(x):
    if x > 0:
        p = a*np.exp(-x**(-a))/x**(1+a)
    else:
        p = 0
    return p

#2) Define scaling constants



#print(res)

#3) Define pmax
#P = [quad(p,0,i)[0] for i in x]


def P(x):
    return quad(p,0,x)[0]

def pmax(x):
    pmax = N*p(x)*(P(x))**(N-1)
    return pmax



#print(pmax(10000))    
#pmaxa = [bn*pmax(i) for i in x]    
#plt.plot(x,pmaxa)

#4) Define density of states

r = np.linspace(0,5*bn,100)


    
def Integrand(x,r):
    return p(x-r)*pmax(x)



rho_a_1  = [bn*quad(Integrand, i , np.inf, args=(i,))[0] for i in r]

def pf(x):
    if x > 0:
        pf = a*np.exp(-x**(-a))/x**(1+a)
    else:
        pf = 0
    return pf


pf_a = [pf(i/bn) for i in r]
## Plotting the results
plt.figure(figsize=(12, 8))
plt.grid()


plt.plot(r/bn, rho_a_1, label = "100000")
plt.plot(r/bn, pf_a, label = "limit")

plt.ylabel(r'  $ \ \  b_{N}\overline{\rho (r, N)}  \ \  $ ',fontsize=16)
plt.xlabel(r'  $\ \ r/b_{N}   \ \  $ ', fontsize=16)
plt.legend(fontsize=20)

plt.savefig("Part1_2.png")


    
## we need to define pmax as a function
```
