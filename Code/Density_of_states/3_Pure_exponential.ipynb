```python
from openpyxl import load_workbook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style # import style module
from scipy import stats
import math
from scipy.integrate import quad

d = 1
N = 10
an = (math.log(N))
bn = 1

x = np.linspace(0.000000000001, 10*bn , 100 )

#1) Define parent function
def p(x):
    if x > 0:
        p = np.exp(-x**(d))
    else:
        p = 0
    return p

#p_a = [p(i) for i in x]
#plt.plot(x, p_a)



def P(x):
    return quad(p,0,x)[0]

def pmax(x):
    pmax = N*p(x)*(P(x))**(N-1)
    return pmax


r = np.linspace(0,10,100)

r2 = np.linspace(-5,5,100)
    
def Integrand(x,r):
    return p(x-r)*pmax(x)



def pf(r):
    pf = np.exp(r)*(1- (1 + np.exp(-r) )*np.exp(-np.exp(-r)))
    return pf


pf_a = [pf(i/bn) for i in r2]


## Plotting everything
plt.figure(figsize=(12, 8))
plt.grid()


rho_a_1  = [quad(Integrand, i , np.inf, args=(i,))[0] for i in r]

plt.plot(r2, pf_a, label = "limit")
plt.plot((r-an), rho_a_1, label = "10")

plt.ylabel(r'  $ \ \  \overline{\rho (r, N)}  \ \  $ ',fontsize=16)
plt.xlabel(r'  $\ \ (r-a_{n})  \ \  $ ', fontsize=16)
plt.legend(fontsize=20)

plt.savefig("Part4_1.png")

```
